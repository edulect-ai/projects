## Named Entity Recognition (NER)

**Named Entity Recognition (NER)** is a sub-task of information extraction in Natural Language Processing (NLP). It involves scanning a text, locating proper nouns or key informational phrases, and classifying them into pre-defined categories (like People, Organizations, Locations, Dates, or Monetary Values).

[Image of Named Entity Recognition text annotation example]

### Why is NER Important?
While POS tagging tells you that a word is a "Noun," NER tells you exactly *what kind* of real-world object that noun represents. 

If you are analyzing thousands of news articles, NER allows you to automatically extract all the companies mentioned, the locations they operate in, and the key executives involved, without having to read a single article yourself.

### Common NER Categories
Different NLP models use slightly different tags, but the standard categories include:

| Entity Tag | Meaning | Examples |
| :--- | :--- | :--- |
| `PERSON` / `PER` | People, including fictional. | Dr. Aris Thorne, Elon Musk, Batman |
| `ORG` | Companies, agencies, institutions. | Google, FBI, United Nations |
| `GPE` | Geopolitical Entities (countries, cities). | London, India, Texas |
| `LOC` | Non-GPE locations, mountain ranges, bodies of water. | The Alps, Pacific Ocean, ancient forest |
| `DATE` | Absolute or relative dates or periods. | summer solstice, 1999, next Tuesday |
| `MONEY` | Monetary values, including unit. | $1 billion, 50 euros |

### Example in Action

**Original Sentence:**
"Tim Cook announced that Apple will open a new campus in Austin next January."

**NER Extracted Entities:**
* **Tim Cook** -> `PERSON`
* **Apple** -> `ORG`
* **Austin** -> `GPE`
* **next January** -> `DATE`



```python
sentence = "Tim Cook announced that Apple will open a new campus in Austin next January."
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
pos_tag = nltk.pos_tag(words)
```

```python
import nltk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```
