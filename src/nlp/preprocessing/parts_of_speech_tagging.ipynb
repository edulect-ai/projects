## Part-of-Speech (POS) Tagging

**Part-of-Speech (POS) Tagging** is the process of reading text and assigning a grammatical category (like noun, verb, adjective, or adverb) to each individual word (token) based on its definition and its context within the sentence.

<img src="../images/pos_tagging.webp" alt="HTML5 Logo" width="380%">

### Why is POS Tagging Important?
Words can have completely different meanings depending on how they are used. POS tagging helps the machine learning model understand the actual *context* of a word:

1. **Word Sense Disambiguation:** * "I want to **book** a flight." -> *book* is a Verb (`VB`).
   * "I read a good **book**." -> *book* is a Noun (`NN`).
2. **Lemmatization:** As we saw earlier, a lemmatizer needs to know if a word is a noun or a verb to accurately find its correct dictionary root.
3. **Named Entity Recognition (NER):** Helps identify people, places, and organizations (which are usually Proper Nouns).

### The Penn Treebank Tagset
Most standard NLP libraries (like NLTK) use a widely accepted set of tags called the **Penn Treebank tagset**. Here are the most common ones you will encounter:

| Tag | Meaning | Example Words |
| :--- | :--- | :--- |
| `NN` | Noun, singular | laboratory, forest, star |
| `NNS` | Noun, plural | vials, windows, trees |
| `VB` | Verb, base form | run, eat, jump |
| `VBD` | Verb, past tense | whispered, danced, buried |
| `JJ` | Adjective | emerald, ancient, reclusive |
| `RB` | Adverb | quickly, silently, very |
| `DT` | Determiner | the, a, an |

### Example in Action

Let's look at a short sentence from our earlier corpus:

**Original Sentence:**
"The reclusive scientist whispered."

**POS Tagged Output:**
`[('The', 'DT'), ('reclusive', 'JJ'), ('scientist', 'NN'), ('whispered', 'VBD'), ('.', '.')]`

 

```python
corpus = """The quick brown fox jumps over the lazy dog is an English-language pangram – a sentence that contains all 26 letters of the English alphabet. The phrase is commonly used for touch-typing practice, testing typewriters and computer keyboards, displaying examples of fonts, and other applications involving text where the use of all letters in the alphabet is desired."""
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
nltk.download('averaged_perceptron_tagger_eng')
import nltk
```

```python
sent_tokens = sent_tokenize(corpus)

for i in range(len(sent_tokens)):
    word_token = word_tokenize(sent_tokens[i])
    word_token = [word for word in word_token if word not in set(stopwords.words('english'))]
    tagged_words = nltk.pos_tag(word_token)
    print(tagged_words)
```

```python
words = corpus.split()
print(nltk.pos_tag(words))
```
