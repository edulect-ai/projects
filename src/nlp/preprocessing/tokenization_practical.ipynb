```python
corpus = """The quick brown fox jumps over the lazy dog. It is a classic pangram used in typography. 
However, foxes are actually quite agile animals!"""
```

```python
### Tokenization
### paragraph to sentence tokenization
import nltk
nltk.download('punkt_tab') # a pre-trained model for sentence tokenization
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
print(documents)
```

```python
### paragrah into words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
### sentences to word
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize # also tokenizes word having '  ex gamer's

wordpunct_tokenize(corpus)
```

```python
## Tree bank word tokenizer
from nltk.tokenize import TreebankWordTokenizer # full stop will be not treated as seperate word, only with last full stop it will be considered as seperate word
```

```python
tokenizer = TreebankWordTokenizer() 
tokenizer.tokenize(corpus)
```
