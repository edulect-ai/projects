# Natural Language Processing Concepts

## Terminologies
<hr>
### Corpus
A **corpus** is a large collection of texts or documents used for linguistic analysis and machine learning. 
It serves as the raw material for training NLP models and understanding language patterns.

### Documents
**Documents** are individual text units within a corpus. Each document can be a sentence, paragraph, article, or any text segment.
Documents are the basic building blocks that make up a corpus.

### Vocabulary
A **vocabulary** (or lexicon) is the set of all unique words present in a corpus.
It represents the dictionary of terms that appear across all documents.

## Words
**Words** are the individual tokens or terms that make up documents. They are the smallest meaningful units of text.
Words are combined to form the vocabulary and make up the content of each document.

#### Relationships:
- Corpus = Collection of Documents
- Documents = Collections of Words
- Vocabulary = Set of all unique Words across all Documents
<hr>



## Tokenization

**Tokenization** is the process of breaking a stream of text up into words, phrases, symbols, or other meaningful elements called *tokens*.  
It’s often the first step in natural language processing and text analysis.

- Tokens can be words, subwords, punctuation marks, or even characters.
- A tokenizer defines the rules for splitting text (e.g. whitespace, punctuation).
- The resulting tokens form the basis for further tasks such as parsing, indexing, or embedding.

Proper tokenization helps models understand the structure and content of the text.

### NLP Tokenization Example

Tokenization is the foundational step in Natural Language Processing (NLP) where continuous text is broken down into smaller, manageable units. 

#### 1. The Corpus (Input)
Imagine a Markdown file named `learning_nlp.md` with the following raw content:

```text
# NLP Basics
Tokenization is the first step. It helps models understand text!
```

#### 2. Level 1: Sentence/Document Tokenization
At this stage, the "Corpus" is broken down into individual sentences or structural segments. In Markdown, we often treat headers and paragraph lines as distinct units.

* **Segment 1:** `# NLP Basics`
* **Segment 2:** `Tokenization is the first step.`
* **Segment 3:** `It helps models understand text!`

#### 3. Level 2: Word Tokenization
Each sentence from the previous step is now broken down into individual units (tokens). Notice how punctuation and Markdown syntax are often treated as separate tokens.

#### Token Breakdown

| Segment ID | Tokens |
| :--- | :--- |
| **S1** | `["#", "NLP", "Basics"]` |
| **S2** | `["Tokenization", "is", "the", "first", "step", "."]` |
| **S3** | `["It", "helps", "models", "understand", "text", "!"]` |

#### 4. Summary of Hierarchy

1. **Corpus:** The entire `.md` file.
2. **Sentences:** Segmenting by boundaries like `.`, `!`, or newlines.
3. **Words:** Splitting by whitespace and isolating punctuation.

### NLP Vocabulary Example

After tokenizing a corpus into words, the next crucial step in Natural Language Processing is building a **Vocabulary**. A vocabulary is a mathematical dictionary containing the unique set of all tokens found in your dataset, mapped to numerical IDs. 

Machine learning models cannot read text; they only understand numbers. The vocabulary is the bridge between text and numbers.

#### 1. The Extracted Tokens
Let's take the tokens extracted from our previous `learning_nlp.md` example. 

**Raw Token List:**
`["#", "NLP", "Basics", "Tokenization", "is", "the", "first", "step", ".", "It", "helps", "models", "understand", "text", "!"]`

#### 2. Creating the Unique Set
To build the vocabulary, we filter the token list down to **unique items only**. *(Note: In standard pipelines, we typically lowercase all text first so that "The" and "the" aren't counted as two different words. For this example, we will keep the original casing).*

Because our example was very short, every token only appeared once.

#### 3. The Vocabulary Mapping (Token to ID)
Next, we assign a unique integer (ID) to every token. We also typically reserve the first few IDs for "Special Tokens" that help the model manage the data.

#### Vocabulary Table

| ID | Token | Description |
| :--- | :--- | :--- |
| `0` | `<PAD>` | **Padding:** Used to make all sentences the same length. |
| `1` | `<UNK>` | **Unknown:** Used if the model encounters a brand new word later. |
| `2` | `#` | Markdown header syntax. |
| `3` | `NLP` | |
| `4` | `Basics` | |
| `5` | `Tokenization` | |
| `6` | `is` | |
| `7` | `the` | |
| `8` | `first` | |
| `9` | `step` | |
| `10` | `.` | Punctuation. |
| `11` | `It` | |
| `12` | `helps` | |
| `13` | `models` | |
| `14` | `understand`| |
| `15` | `text` | |
| `16` | `!` | Punctuation. |

#### 4. Encoding Text (Vectorization)
Now that we have a vocabulary, we can translate our text into a format the computer can process. This is called encoding.

**Original Sentence:**
`"Tokenization is the first step."`

**Tokenized Array:**
`["Tokenization", "is", "the", "first", "step", "."]`

**Numerical Representation (Using Vocab IDs):**
`[5, 6, 7, 8, 9, 10]`
