### Download dataset


```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/sms-spam-collection-dataset")

print("Path to dataset files:", path)
```

### Load dataset and prepare DataFrame


```python
import pandas as pd
import os

path2 = os.path.join(path, 'spam.csv')
print(path2)
messages = pd.read_csv(os.path.join(path, 'spam.csv'), usecols=['v1','v2'], encoding='latin1').rename(columns={'v1': 'label', 'v2': 'message'})

messages = messages[messages['label'] != 'v1']


```

### Import regular expressions and NLTK stopwords


```python
import re
import nltk
nltk.download('stopwords')
```

### Import stopwords and Porter stemmer


```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

### Preprocess SMS messages into a stemmed corpus


```python
ps = PorterStemmer()
corpus = []
for i in range (0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i]) # replace non-alphabetic with space

    review = review.lower()

    review = review.split()

    review = [ ps.stem(word) for word in review if word not in stopwords.words("english")]

    review = ' '.join(review)

    corpus.append(review)

print(corpus[:5])  # print first 5 to check
```

### Convert corpus to bag-of-words vectors


```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=False, max_features=500)

X = cv.fit_transform(corpus).toarray()
print(X)
```

```python
##

cvBigram = CountVectorizer(binary=True, max_features=500, ngram_range=(1,2))

X1 = cvBigram.fit_transform(corpus).toarray()

print(cvBigram.vocabulary_)
```

In the above cell output we will she the word with the index, for example word with index 1 has the highest occurance.
