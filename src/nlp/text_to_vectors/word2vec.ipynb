# Understanding Word Embeddings

## What is a Word Embedding?
A **Word Embedding** is a technique in Natural Language Processing (NLP) where words are represented as dense vectors (lists of real numbers) in a continuous vector space. 

Unlike the basic counting methods (like Bag of Words or TF-IDF) which create massive, sparse matrices mostly filled with zeros, word embeddings compress the English language into a smaller, dense space. Most importantly, **they capture the semantic meaning of words**. In a good word embedding model, words with similar meanings (like "king" and "queen", or "fast" and "quick") will be located physically close to each other in the mathematical vector space.

---

## The Word Embedding Family Tree

Word embeddings are broadly categorized into two main types based on how they generate their mathematical vectors:

```text
Word Embeddings
│
├── 1. Count-Based Methods (Frequency & Co-occurrence)
│   ├── Bag of Words (BoW): Counts raw word frequency.
│   ├── TF-IDF: Weighs word frequency against global rarity.
│   └── Co-occurrence Matrix: Counts how often words appear next to each other.
│
└── 2. Deep Learning-Based Methods (Predictive Models)
    ├── Word2Vec: Uses a neural network to predict words based on surrounding context.
    ├── GloVe: Combines global matrix factorization with local context windows.
    ├── FastText: Breaks words into sub-words (n-grams) to understand prefixes/suffixes.
    └── Contextual Models (BERT, GPT): Advanced models where the embedding changes based on the sentence context (e.g., "bank" of a river vs. "bank" for money).

# Understanding Word2Vec

## What is Word2Vec?
**Word2Vec** is a famous deep learning technique created by researchers at Google in 2013. It is used to generate high-quality **word embeddings** (dense vector representations of words). 

The entire model is built on a famous linguistic theory by John Rupert Firth: *"You shall know a word by the company it keeps."* Instead of counting how many times words appear in a document (like TF-IDF), Word2Vec trains a shallow, two-layer neural network to **predict** words based on their surrounding context. 


# Word2Vec Feature Vectors: A Conceptual Example

To understand how Word2Vec represents words, we have to look at **Latent Dimensions**. 

Unlike Bag of Words or TF-IDF, the columns in Word2Vec are no longer specific words or documents. Instead, they are hidden, abstract features discovered by the neural network on its own. The algorithm doesn't explicitly name these columns, but for human understanding, we assign conceptual labels to them to show how the math represents meaning.

*(Note: Values typically range between -1.0 and 1.0, where numbers closer to 1 or -1 show a strong relationship, and numbers near 0 show no relationship).*

## The Feature Vector Matrix

| Word | Dim 1: "Gender" (Masculine) | Dim 2: "Royalty" | Dim 3: "Age" (Adult) | Dim 4: "Food" (Edible) |
| :--- | :--- | :--- | :--- | :--- |
| **King** | +0.95 | +0.93 | +0.70 | -0.85 |
| **Queen** | -0.92 | +0.95 | +0.68 | -0.88 |
| **Man** | +0.96 | -0.80 | +0.65 | -0.80 |
| **Woman** | -0.95 | -0.82 | +0.66 | -0.82 |
| **Boy** | +0.94 | -0.85 | -0.75 | -0.78 |
| **Apple** | +0.02 | -0.95 | -0.10 | +0.98 |

---

## How to Read the Vector Data

If we pull the vector for the word **"King"**, it looks like this: `[0.95, 0.93, 0.70, -0.85]`

Here is how the neural network mathematically understands a King based on that vector:
* **Gender (+0.95):** It is highly masculine.
* **Royalty (+0.93):** It is highly associated with power and royalty.
* **Age (+0.70):** It is associated with adults, not children.
* **Food (-0.85):** It is strongly anti-associated with food (you don't eat a king).

Look at the vector for **"Apple"**. It scores near zero (`0.02`) for gender because apples don't have genders. It scores very negatively for royalty, but it scores almost a perfect `+1.0` in the food dimension.

---

## The Famous Vector Math Example

Because these definitions are just numbers on a grid, you can literally do math with the English language. Let's look at the famous `King - Man + Woman` equation using the specific data from our table.

1. **Start with King:** `[0.95, 0.93, 0.70, -0.85]`
2. **Subtract "Man":** We are essentially telling the computer, *"Take the concept of a King, and remove the male gender."*
   * `Gender: +0.95 - (+0.96) = -0.01` (Effectively 0)
   * `Royalty: +0.93 - (-0.80) = +1.73` (The royalty score stays very high)
3. **Add "Woman":** Now we tell the computer, *"Add the female gender back into this royal concept."*
   * `Gender: -0.01 + (-0.95) = -0.96` 

If we calculate this for all the columns, our resulting "math" vector looks extremely close to `[-0.92, +0.95, +0.68, -0.88]`. 

If the computer searches its database to find which single word's vector matches those coordinates the closest, it will return the word **"Queen"**!
