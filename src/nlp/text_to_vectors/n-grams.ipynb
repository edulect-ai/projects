# Understanding N-grams and Bigrams

An **N-gram** is a contiguous sequence of *n* items from a given sample of text. By grouping adjacent words together, N-gram models can capture local context, word order, and semantic meaning that standard unigram (single-word) models completely miss.

* **N = 1 (Unigram):** Bag of words (e.g., "climate", "is")
* **N = 2 (Bigram):** Word pairs (e.g., "climate is", "is good")
* **N = 3 (Trigram):** Three-word chunks (e.g., "the climate is")

## The Problem with Single Words (Unigrams)

Consider these two sentences:
1. "the climate is good"
2. "the climate is not good"

A standard unigram model sees the word "good" in both sentences and scores them as highly similar. It loses the crucial context that the word "not" modifies "good."

## The Bigram Solution

Let's process the exact same sentences using **Bigrams** (2-grams).

### Step 1: Extract Bigrams and Build Vocabulary
We break both sentences into overlapping pairs of words to create our master vocabulary.

**Master Vocabulary:**
1. the climate
2. climate is
3. is good
4. is not
5. not good

### Step 2: Vectorization
We score each sentence based on the presence of our vocabulary bigrams.

**Sentence 1:** "the climate is good"
* the climate: 1
* climate is: 1
* is good: 1
* is not: 0
* not good: 0
* **Vector:** `[1, 1, 1, 0, 0]`

**Sentence 2:** "the climate is not good"
* the climate: 1
* climate is: 1
* is good: 0
* is not: 1
* not good: 1
* **Vector:** `[1, 1, 0, 1, 1]`

## Summary Matrix

| Document | the climate | climate is | is good | is not | not good |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **Sentence 1** | 1 | 1 | 1 | 0 | 0 |
| **Sentence 2** | 1 | 1 | 0 | 1 | 1 |

**Conclusion:** By looking at the vectors, you can see how bigrams successfully capture negation. Sentence 1 triggers the `is good` feature, while Sentence 2 safely bypasses it, triggering `is not` and `not good` instead. The mathematical vectors now accurately reflect the opposite meanings of the sentences.
