```python
## try , except , else block
try:
    num = int(input("Enter a number"))
    result = 10/num
except ValueError:
    print("This is not valid number")
except ZeroDivisionError:
    print("Enter denominator not equal to 0")
except Exception as ex:
    print(ex)
else:
    print(result) # if exception is not thrown else block will get executed
```

```python
## try , except , finally
try:
    num = int(input("Enter a number"))
    result = 10/num
except ValueError:
    print("This is not valid number")
except ZeroDivisionError:
    print("Enter denominator not equal to 0")
except Exception as ex:
    print(ex)
else:
    print(result) # if exception is not thrown else block will get executed
finally:
    print("Excution completed") # always get executed even if there is error
```
