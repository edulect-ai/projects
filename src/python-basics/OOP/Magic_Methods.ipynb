```python
# Magic methods (dunder methods) in Python are special methods that start and end with double underscores.
# They allow customization of class behavior. Here are some common examples:

class Example:
    def __init__(self, value):
        # __init__: Initializes the object when created.
        self.value = value
    
    def __str__(self):
        # __str__: Defines the string representation of the object (used by str() and print()).
        return f"Example with value: {self.value}"
    
    def __repr__(self):
        # __repr__: Provides an unambiguous string representation (used by repr()).
        return f"Example({self.value!r})"
    
    def __add__(self, other):
        # __add__: Defines behavior for the + operator.
        if isinstance(other, Example):
            return Example(self.value + other.value)
        return NotImplemented
    
    def __len__(self):
        # __len__: Defines behavior for len() function.
        return len(str(self.value))
    
    def __eq__(self, other):
        # __eq__: Defines behavior for == comparison.
        return isinstance(other, Example) and self.value == other.value

# Example usage:
obj1 = Example(10)
obj2 = Example(20)
print(obj1)  # Uses __str__
print(repr(obj1))  # Uses __repr__
obj3 = obj1 + obj2  # Uses __add__
print(len(obj1))  # Uses __len__
print(obj1 == obj2)  # Uses __eq__
```

```python
### Basic magic methods
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self): ## override default str which prints the memory location where object is created.
        return f"Person create with name {self.name}"
    
    def __repr__(self): # Returns a string that can recreate the object
        return f"Person (name= {self.name})"

person = Person("Hrishi", 25)
print(person)
repr(person)
```

```python
### operator overriding

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    

## create object of the vector class
v1 = Vector(2,3)
v2 = Vector(4,5)

print(v1+v2)
```
