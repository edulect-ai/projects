```python
## Base class

class Animal:
    def speak(self):
        return "Sound of animal"
    
class Dog(Animal):
    def speak(self):
        return "Bark" ## override the speak method Animal
    
class Cat(Animal):
    def speak(self):
        return "Meow!!" ## Override the speak method of animal

def animal_speak(animal):
    print(f"Animal speak {animal.speak()}")
dog = Dog()
dog.speak()
animal_speak(dog)
```

```python
### Polymorphism with functions and methods

## base class
class Shape:
    def area(self):
        return "The area of the figure"
    
## Derieved class 1
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
    
## Derieved class 2
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius**2
    
## Function that demostrates polymorphism
def print_area(shape):
    print(f"the area is {shape.area()}")

rectangle = Rectangle(4,5)
circle = Circle(3)

print_area(rectangle)
print_area(circle)
```

```python
## Polymorphism with abstract base class
from abc import ABC, abstractmethod

## Define a abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

## Derieved class 1
class Car(Vehicle):
    def start_engine(self):
        return "car engine start"

## Derieved class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motor Cycle Engine started"

car = Car()
motorcycle = Motorcycle()

## Demo
def start_engine(car):
    print(car.start_engine())

start_engine(car)
start_engine(motorcycle)
```
