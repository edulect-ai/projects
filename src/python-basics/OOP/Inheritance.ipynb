```python
## Inheritence
class Car:
    def __init__(self, windows, doors, engineType):
        self.windows = windows
        self.doors = doors
        self.engineType = engineType

    def drive(self):
        print(f"The person will drive the {self.engineType} car")

class Audi(Car):
    def __init__(self, windows, doors, engineType, isSelfDriving):
        super().__init__(windows, doors, engineType)
        self.is_selfDriving = isSelfDriving
    def self_drive(self):
        print(f"Is self Drive allowed {self.is_selfDriving}")

audi = Audi(4,4,'disel', True)

audi.drive()
audi.self_drive()
```

```python
## Multiple inheritence
## When a class in herits from more than one base class

class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print("Sub class")

class Pet:
    def __init__(self, owner):
        self.owner = owner

class Dog(Animal,Pet):
    def __init__(self,name, owner):
        Animal.__init__(self,name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} say woof"
    
## create an object
dog = Dog("buddy", "Hrihsi")
dog.speak()
print(f"Owner {dog.owner}")
```
