# Tuples in Python
This notebook covers tuple creation, indexing, immutability, packing/unpacking, and nested tuples.

## Accessing elements
Tuples support indexing and slicing just like lists.

## Tuple operations
You can concatenate and repeat tuples; they behave like immutable lists.

```python
# simple tuple literal
number = (1, 2, 5)
print(type(number))  # <class 'tuple'>
```

## Immutability and methods
Tuples cannot be modified in-place but provide `count` and `index`.

## Packing and unpacking
Tuples can be constructed by comma-separated values and unpacked into variables.

## Nested tuples
Tuples can contain other tuples; iterate recursively if needed.

```python
# accessing tuple elements

numbers = (0, 1, 2, 3, 4, 5)

print(numbers[3])

# indexing and slicing operations are similar to list
```

```python
# tuple operations

numbers = (0, 2, 3, 4, 5, 6)
mixed_tuple = ("hello", 1, 'c', 0.5)

concate = numbers + mixed_tuple
print(concate)
```

```python
# repeat tuple (creates a new tuple)
numbers = numbers * 3
print(numbers)
```

```python
# Tuples elements cannot be changed once assigned
# methods that don’t modify the tuple:

numbers = (1, 2, 3, 4, 1, 2, 4, 51, 1, 0, 2, 24, 3)

print(numbers.count(1))  # count occurrences of 1
print(numbers.index(3))  # index of first occurrence of 3
```

```python
packed_tuple = 1, "hello", 3.5
print(packed_tuple)

a, b, c = packed_tuple
print(f"{a}, {b}, {c}")

# unpacking with * to collect remaining items
numbers = (1,2,3,4,5,6)
first, *middle, Last = numbers
print(f"{first}, {middle}, {Last}")
```

```python
nested = ((1,2,3,4), (4,5,6,7), (8,9,2,3))

# access the elements
print(nested[0])
print(nested[1][2])

# iteration over nested tuples
for sub_tuple in nested:
    for item in sub_tuple:
        print(item, end=" ")
```
