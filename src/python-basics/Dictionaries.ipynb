# Dictionaries in Python
This notebook covers the basics of creating and using dictionaries.  Each section includes explanatory comments and examples.

```python
# create an empty dictionary using literal syntax
# every dictionary is an unordered collection of key:value pairs

dict_1 = {}

print(type(dict_1))  # confirm the type is dict

# the constructor form is equivalent
# ensure built-in name `dict` hasn't been shadowed

dict_2 = dict()
print(dict_2)  # empty dictionary created by calling the type
```

```python
# demonstrate that keys are unique and can store various types
student = {"name": "hrishi", "age": 32, "grade": 24}
print(student)  # prints the dictionary with three key:value pairs
```

```python
# show what happens when duplicate keys are used
# the later value wins and no error is raised
student = {"name": "hrishikesh", "age":32, "name":24}
print(student)  # name will be 24
```

## Accessing dictionary elements
This section demonstrates how to look up values, add new entries, and remove existing ones.

## Common dictionary methods
Dictionaries provide view objects that reflect the current contents of the mapping.

## Shallow copy
Assigning one dictionary to another name doesn\'t copy it; both names refer to the same object. A shallow copy duplicates only the outer structure.

### Deep copy example
A nested container requires a deep copy if you want completely independent clones of every element.

## Iterating over dictionaries
You can loop over keys, values, or both simultaneously using methods that return iterable views.

```python
student = {"name": "hrishi", "age": 32, "grade": "A"}

# retrieve a value by key (raises KeyError if missing)
student['grade']

# access using the safe `get` method, with optional default
student.get('age')

# add or update a key
student["address"] = "India"
print(student)

# delete a key
if 'address' in student:
    del student['address']
print(student)
```

```python
# get all keys (returns a view, not a list)
keys = student.keys()
print(keys)

# get all values
values = student.values()
print(values)

# get all key/value pairs as (key, value) tuples
pairs = student.items()
print(pairs)
```

```python
# original dictionary
student_copy = student

# both names refer to same object; modifying one affects the other
print(student)
print(student_copy)

student_copy["age"] = 53
print(student)        # student shows updated age
print(student_copy)

# use the `copy` method to create a shallow copy
student_copy = student.copy()
student_copy["name"] = "Rak"
print(student)        # original still has old name
print(student_copy)
```

```python
# "shallow copy" only duplicates the outer object; inner objects remain shared
import copy

orig = {'a': [1, 2], 'b': {'x': 10}}
shallow = orig.copy()          # or dict(orig), or orig[:] for lists
deep = copy.deepcopy(orig)

# mutating the nested list affects the shallow copy but not the deep copy
orig['a'].append(3)
print(shallow['a'])   # -> [1, 2, 3]    (shares same list)
print(deep['a'])      # -> [1, 2]
```

```python
# Use loops to iterate over dictionaries
student = {"name": "hrishi", "age": 32, "grade": "A"}

# Iterating over keys
for key in student.keys():
    print(key)

# Iterate over values
for value in student.values():
    print(value)

# Iterate over key value pairs
for k,v in student.items():
    print(k,v)
```

## Nested dictionaries
Dictionaries can contain other dictionaries (or any object) as values.

### Iterating over nested dictionaries
Use nested loops to walk internal mappings.

## Dictionary comprehension
Construct dictionaries concisely using comprehension syntax, optionally with filters.

## Practical example: frequency count
Use a dictionary to tally occurrences in a sequence.

## Merging dictionaries
Combine multiple mappings; later keys override earlier ones.

```python
students = {
    "student1": {
        "name": "hrishi",
        "age":32,
    },

    "student2": {
        "name": "lucky",
        "age":35,
    }
}

# print the full nested structure
print(students)


# access nested fields using chained indexing
print(students["student1"]["name"])
```

```python
# iterating over nested dict
for s_id, s_info in students.items():
    print(s_id, s_info)
    for k, v in s_info.items():
        print(k, v)
```

```python
# basic comprehension
squares = {x: x**2 for x in range(5) }
print(squares)

# conditional comprehension (only even numbers)
squares = {x: x**2 for x in range(5) if x % 2 ==0}
print(squares)
```

```python
numbers = [1,2,2,3,3,3,1,4,5]

freq = {}  # avoid shadowing built-in name 'dict'
for number in numbers:
    # get current count or 0 if missing, then add one
    freq[number] = freq.get(number, 0) + 1
print(freq)
```

```python
dict1 = { "a":1, "b":2}
dict2 = {"b":3, "c":4}

merge_dict = {**dict1,**dict2}  # unpacking merge
print(merge_dict)
```
