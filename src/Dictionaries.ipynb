```python
## accessing dictionary elements
student = {"name": "hrishi", "age": 32, "grade": "A"}

student['grade']

## access using default get method
student.get('age')

## add new key
student["address"] = "India"
print(student)

## delete a key
del student['address']
print(student)

```

```python
## Common dictionaries method

## get all keys
keys = student.keys()
print(keys)

## get all values
values = student.values()
print(values)

## get all key value pairs
pair = student.items()
print(pair)
```

```python
## Shallow copy
student_copy = student

print(student)
print(student_copy)

student_copy["age"] = 53

print(student)
print(student_copy)

student_copy = student.copy()
student_copy["name"] = "Rak"
print(student)
print(student_copy)
```

```python
#“shallow copy” only duplicates the outer object; 
# the inner pieces are still references to the same objects.
# hece we can do deep copy as below
import copy

orig = {'a': [1, 2], 'b': {'x': 10}}
shallow = orig.copy()          # or dict(orig), or orig[:] for lists
deep = copy.deepcopy(orig)

orig['a'].append(3)
print(shallow['a'])   # -> [1, 2, 3]    (shares same list)
print(deep['a'])    
```

```python
## Iteration

## Use loops to iterate over dictionaries
student = {"name": "hrishi", "age": 32, "grade": "A"}

## Iterating over keye
for key in student.keys():
    print(key)

## Iterate over values
for value in student.values():
    print(value)

## Iterate over key value pair
for k,v in student.items():
    print(k,v)
```

```python
## Nested dictionaries
students = {
    "student1": {
        "name": "hrishi",
        "age":32,
    },

    "student2": {
        "name": "lucky",
        "age":35,
    }
}

print(students)
print(students["student1"]["name"])
```

```python
# iterating over nested dict
for s_id,s_info in students.items():
    print(s_id, s_info)
    for k,v in s_info.items():
        print(k,v)
```

```python
## Dictionary comprehension
squares = {x: x**2 for x in range(5) }
print(squares)

## Conditional Dictionary comprehension
squares = {x: x**2 for x in range(5) if x % 2 ==0}
print(squares)
```

```python
## Practical example: Count the frequency of elements in a list

numbers = [1,2,2,3,3,3,1,4,5]

dict = {}

for number in numbers:
    if(not dict.get(number)):
        dict[number] = 1
    else:
        dict[number] += 1
print(dict)
```

```python
## Merge dictionaries into 1
dict1 = { "a":1, "b":2}
dict2 = {"b":3, "c":4}

merge_dict = {**dict1,**dict2}
print(merge_dict)
```
