```python
#accessing tuple elements

numbers = (0 ,1, 2,3, 4,5)

numbers[3]

# indexing and slicing operations are similiar to list
```

```python
# tuple operations

numbers = (0, 2, 3,4,5,6)
mixed_tuple = ("hello", 1, 'c', 0.5)

concate = numbers + mixed_tuple
print(concate)
```

```python
# repeat tuple 
numbers = numbers * 3
print(numbers)
```

```python
# immutable nature of tuples

# Tuples elements cannot be changed once assigned

# Tuples method

numbers = (1, 2,3, 4,1,2,4,51 , 1,0, 2,24, 3)

print(numbers.count(1)) # count number of 1
print(numbers.index(3)) # index of 3 's first occuerance
```

```python
## Packing and unpacking tuple

packed_tuple = 1, "hello", 3.5
print(packed_tuple)

a,b,c = packed_tuple
print(f"{a}, {b}, {c}")

## unpacking with *
numbers = (1,2,3,4,5,6)
first,*middle,Last = numbers
print(f"{first}, {middle}, {Last}")
```

```python
## Nested tuples

nested = ((1,2,3,4), (4,5,6,7), (8,9,2,3))

## access the elements
print(nested[0])
print(nested[1][2])


## iteration over nested tuples

for sub_tuple in nested:
    for item in sub_tuple:
        print(item,end=" ")
```
