```python
### read a whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# read line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) # Removes the new line character
```

```python
# over writing a file
with open("example.txt", 'w') as file:
    file.write("This is a good example.")
    file.write("This is a new line")
```

```python
# write file without over writing

with open("example.txt", 'a') as file:
    file.write("This is a new line")
```

```python
# writing list of lines to a file
line  = ["this is first line \n", "this is second line \n", "this is third line \n"]

with open("example.txt", 'a') as file:
    file.writelines(line)
```

```python
## how to work with binary files
# Writing bytes to a binary file
with open('example.bin', 'wb') as file:
    file.write(b'\x00\x01\x02\x03\x04\x05')  # Example: writing 6 bytes
```

```python
## Read from a file and write to a destination file
with open("example.txt", 'r') as file:
    content = file.read()

with open("dest.txt", 'w') as file:
    file.write(content)
```

```python
# Read file and count number of lines and charaters

def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        words = sum(len(line.split()) for line in lines)
        char = sum(len(line) for line in lines)
        print(len(lines))
        print(words)
        print(char)

count_text_file('example.txt')
```

```python
# writing and then reading a file

with open('example2.txt', 'w+') as file:  # w+ is for read and write if file is not there it will be created as overwritten
    file.write("this is line 1\n")
    file.write("this is line 2\n")

    file.seek(0) # move file cursor to the begining
    content = file.read()
    print(content)
```

```python
# create a new directory
import os
new_dir = "package_new"
os.mkdir(new_dir)
```

```python

## Listing files and dirs
items = os.listdir(".")
print(items)
```

```python
## joining paths
dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(dir_name,file_name)
print(full_path)
abs_path = os.path.join(os.getcwd(), dir_name,file_name)
print(abs_path)
```

```python
## check if path is there
path = "example.txt"
if os.path.exists(path):
    print(f"the path {path} exist")
```

```python
## check if path is file is directory
if os.path.isfile(path):
    print("is a file")
elif os.path.isdir(path):
    print("is a directory")
```

```python
## getting the absolute path

relative_path = "example.txt"
abs_path = os.path.abspath(relative_path)
print(abs_path)
```
